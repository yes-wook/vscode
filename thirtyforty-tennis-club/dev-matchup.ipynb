```python
import random
from collections import defaultdict

# 참가자 목록
participants = ['정현', '쥬드', '성현', '연주', '파란하늘', '유하', '미지']

# 각 참가자의 경기 횟수를 기록할 딕셔너리
match_count = defaultdict(int)

# 총 경기 수
num_matches = 8

# 대진표 생성 함수
def create_schedule():
    matches = []

    while len(matches) < num_matches:
        # 현재 경기 수를 고려한 참가자 목록 생성
        min_count = min(match_count.values()) if match_count else 0
        
        # 덜 들어간 사람들을 우선으로 고려
        available_participants = [p for p in participants if match_count[p] <= min_count + 1]

        # 4명 이상일 때만 경기 생성
        if len(available_participants) >= 4:
            match = random.sample(available_participants, 4)
        else:
            # 모든 참가자를 사용할 수 밖에 없는 경우
            match = random.sample(participants, 4)

        # 선택된 참가자의 경기 횟수를 증가
        for participant in match:
            match_count[participant] += 1

        matches.append(match)

    return matches

# 스케줄 생성
matches = create_schedule()

# 경기 결과 출력
print("Final Match Schedule:")
for i, match in enumerate(matches, 1):
    men_count = sum(1 for p in match if p in ['정현', '쥬드', '성현'])
    women_count = sum(1 for p in match if p in ['연주', '파란하늘', '유하', '미지'])
    print(f"Match {i}: {' '.join(match)} (남자: {men_count}, 여자: {women_count})")

# 각 참가자 별 경기 횟수 출력
print("\nMatch Counts:")
for participant, count in match_count.items():
    print(f"{participant}: {count} matches")

# 전체 경기 수 차이 확인
max_count = max(match_count.values())
min_count = min(match_count.values())
print(f"\nMaximum matches: {max_count}, Minimum matches: {min_count}, Difference: {max_count - min_count}")
```
