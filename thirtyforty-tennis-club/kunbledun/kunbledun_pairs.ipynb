```python
import random
import pandas as pd
from IPython.display import display
import itertools

# 중복된 페어를 방지하는 함수
def create_pair_no_duplicates(group_a, group_b, group_a_name, group_b_name, round, existing_pairs):
    pairings = []
    random.shuffle(group_a)
    random.shuffle(group_b)

    for i in range(min(len(group_a), len(group_b))):
        pair = (group_a[i], group_b[i], f"{group_a_name}-{group_b_name}", round)
        
        # 중복된 페어가 아닌지 확인
        if (group_a[i], group_b[i]) not in existing_pairs and (group_b[i], group_a[i]) not in existing_pairs:  # 역방향으로도 체크
            pairings.append(pair)
            existing_pairs.add((group_a[i], group_b[i]))  # 페어 추가 (정방향)
    
    return pairings


# 혼복이후 남은 사람들을 섞는 함수 - 중복 페어 방지
def create_random_leftover_pairs_no_duplicates(remaining_men_list, round, existing_pairs):
    # 조별로 분류
    group_1 = [p for p in remaining_men_list if p[1] == '1조']
    group_2 = [p for p in remaining_men_list if p[1] == '2조']
    group_3 = [p for p in remaining_men_list if p[1] == '3조']
    
    # 각 그룹 무작위로 섞기
    random.shuffle(group_1)
    random.shuffle(group_2)
    random.shuffle(group_3)
    
    # 서로 다른 조끼리 페어링 (1조-2조, 2조-3조, 3조-1조)
    pair_1 = (group_1[0][0], group_2[0][0], f"{group_1[0][1]}(남)-{group_2[0][1]}(남)", round)
    pair_2 = (group_2[1][0], group_3[0][0], f"{group_2[1][1]}(남)-{group_3[0][1]}(남)", round)
    pair_3 = (group_1[1][0], group_3[1][0], f"{group_1[1][1]}(남)-{group_3[1][1]}(남)", round)
    
    # 중복 페어 확인 및 저장
    for pair in [pair_1, pair_2, pair_3]:
        if (pair[0], pair[1]) not in existing_pairs and (pair[1], pair[0]) not in existing_pairs:  # 역방향으로도 체크
            existing_pairs.add((pair[0], pair[1]))  # 페어 추가 (정방향)
    
    return [pair_1, pair_2, pair_3]


# 메인 코드

for team in ['청팀', '백팀']:
    random.seed(dict_random[team])
    
    # 페어링 결과 리스트
    pairing_results = []
    pairing_results_round_2 = []
    pairing_results_round_4 = []

    # 중복 페어 추적을 위한 리스트
    existing_pairs = set()  # 전역적으로 중복 페어를 추적

    # 남자와 여자 추출
    men_team = df[(df['팀'] == team) & (df['성별'] == '남')]
    women_team = df[(df['팀'] == team) & (df['성별'] == '여')]

    # 남자 조별로 분리
    men_group_1 = men_team[men_team['조'] == '1조']['닉네임'].tolist()
    men_group_2 = men_team[men_team['조'] == '2조']['닉네임'].tolist()
    men_group_3 = men_team[men_team['조'] == '3조']['닉네임'].tolist()

    # 여자 조별로 분리
    women_group_1 = women_team[women_team['조'] == '1조']['닉네임'].tolist()
    women_group_2 = women_team[women_team['조'] == '2조']['닉네임'].tolist()
    women_group_3 = women_team[women_team['조'] == '3조']['닉네임'].tolist()

    ######################################################################
    # 여자끼리의 페어
    pairing_results += create_pair_no_duplicates(women_group_2, women_group_3, "2조(여)", "3조(여)", 1, existing_pairs)
    pairing_results += create_pair_no_duplicates(women_group_1, women_group_2, "1조(여)", "2조(여)", 1, existing_pairs)
    pairing_results += create_pair_no_duplicates(women_group_1, women_group_3, "1조(여)", "3조(여)", 1, existing_pairs)

    ######################################################################
    # 남자-여자 페어링
    pairing_results_round_2 += create_pair_no_duplicates(men_group_1, women_group_2, "1조(남)", "2조(여)", 2, existing_pairs)
    pairing_results_round_2 += create_pair_no_duplicates(men_group_2, women_group_3, "2조(남)", "3조(여)", 2, existing_pairs)
    pairing_results_round_2 += create_pair_no_duplicates(men_group_3, women_group_1, "3조(남)", "1조(여)", 2, existing_pairs)

    ######################################################################
    # 남는 남자들의 닉네임과 조 정보를 함께 저장
    remaining_men = men_team[men_team['닉네임'].isin(list(set(men_team['닉네임'].tolist()) - set([p[0] for p in pairing_results_round_2])))]
    remaining_men_list = remaining_men[['닉네임', '조']].values.tolist()  # [[닉네임, 조], ...]

    # 남는 남자끼리 페어링
    remaining_men_pairs_round_2 = create_random_leftover_pairs_no_duplicates(remaining_men_list, 2, existing_pairs)

    pairing_results += pairing_results_round_2
    pairing_results += remaining_men_pairs_round_2

    ######################################################################
    # 남자끼리의 페어
    pairing_results += create_pair_no_duplicates(men_group_1, men_group_3, "1조(남)", "3조(남)", 3, existing_pairs)
    pairing_results += create_pair_no_duplicates(men_group_1, men_group_2, "1조(남)", "2조(남)", 3, existing_pairs)
    pairing_results += create_pair_no_duplicates(men_group_2, men_group_3, "2조(남)", "3조(남)", 3, existing_pairs)

    ######################################################################
    # 남자-여자 페어링
    pairing_results_round_4 += create_pair_no_duplicates(men_group_2, women_group_1, "2조(남)", "1조(여)", 4, existing_pairs)
    pairing_results_round_4 += create_pair_no_duplicates(men_group_3, women_group_2, "3조(남)", "2조(여)", 4, existing_pairs)
    pairing_results_round_4 += create_pair_no_duplicates(men_group_1, women_group_3, "1조(남)", "3조(여)", 4, existing_pairs)

    ######################################################################
    # 남는 남자들의 닉네임과 조 정보를 함께 저장
    remaining_men = men_team[men_team['닉네임'].isin(list(set(men_team['닉네임'].tolist()) - set([p[0] for p in pairing_results_round_4])))]
    remaining_men_list = remaining_men[['닉네임', '조']].values.tolist()

    # 남는 남자끼리 페어링
    remaining_men_pairs_round_4 = create_random_leftover_pairs_no_duplicates(remaining_men_list, 4, existing_pairs)

    pairing_results += pairing_results_round_4
    pairing_results += remaining_men_pairs_round_4

    ######################################################################
    # 결과를 pandas dataframe에 저장
    df_pairs = pd.DataFrame(pairing_results, columns=["선수", "선수2", "조 구분", "라운드"])

    # 인덱스를 1부터 시작하도록 조정
    df_pairs.index = df_pairs.index + 1

    # 인덱스를 '페어번호'라는 이름의 열로 추가
    df_pairs.reset_index(inplace=True)
    df_pairs.rename(columns={'index': '페어번호'}, inplace=True)


    # 페어 유형을 추가하여 구분
    df_pairs['페어 유형'] = df_pairs['조 구분'].apply(
        lambda x: '남복' if '남' in x and '여' not in x else '여복' if '여' in x and '남' not in x else '혼복'
    )

    # 결과 출력
    # display(df_pairs.style.set_table_attributes('style="max-height:300px; overflow:auto;"'))
    df_pairs.to_excel(f'{team}_fair_{dict_random[team]}.xlsx', index=False, engine='openpyxl')
```
